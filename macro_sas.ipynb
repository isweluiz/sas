* - Construa uma macro em linguagem SAS que recebe os parametros

- INPUT_DATA,  que apresenta uma tabela sas qualquer

- VAR_CHAR, que representa uma lista de variáveis categoricas que devem estar na tabela definida em input_data

- VAR_NUM, que representa uma lista de variáveis númericas que devem estar na tabela definida em input_data
